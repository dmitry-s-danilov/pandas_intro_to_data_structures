```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create series

index_list = list("abcdefgh") # set index list

array_length = len(index_list)             # set array length and
data_array = np.random.randn(array_length) # initialize data ndarray

data_series = pd.Series(data_array, index_list) # create data series
                                                # from data ndarray providing index list

# print

print(
    f'index {type(index_list)}:',
    f'{index_list}',
    '',    
    f'data {type(data_array)}:',
    f'{data_array}',
    '',    
    f'series {type(data_series)}:',
    f'{data_series}',
    sep='\n'
)

# plot

plt.plot(data_series, 'o:')
plt.show()
```

```python
# apply methods

data_series_min = data_series.min()       # minimum
data_series_max = data_series.max()       # maximum
data_series_mean = data_series.mean()     # mean
data_series_median = data_series.median() # median

# print

print('series <statistical methods>:')

for index, data_value in enumerate(data_series):
    print(f'{index_list[index]}{data_value:10.6f}', end='')
    if data_value == data_series_min:
        print(' min')
    elif data_value == data_series_max:
        print(' max')
    else:
        print()

print(
    '',
    f'mean:{data_series_mean:10.6f}',
    f'median:{data_series_median:10.6f}',
    sep='\n'
)

# plot 

plt.plot(data_series, 'o:') # plot series

axis_0_tuple = (0, len(data_series) - 1) # set x-positions

axis_1_tuple = (data_series_mean, data_series_mean) # set y-positions and
plt.plot(axis_0_tuple, axis_1_tuple, '.-')          # plot mean line

axis_1_tuple = (data_series_median, data_series_median) # set y-positions and
plt.plot(axis_0_tuple, axis_1_tuple, '.-')              # plot median line

plt.show()
```

```python
# apply conditions

data_series_1_1 = data_series[data_series > data_series_median] # more than median
data_series_1_2 = data_series[data_series < data_series_median] # less than median

# print

print(
    
    f'series <more than median condition>:',
    f'{data_series_1_1}',
    '',
    f'series <less than median condition>:',
    f'{data_series_1_2}',
    sep='\n'
)

# plot

plt.plot(
    data_series, '--',
    data_series_1_1, 'o:',
    data_series_1_2, 'o:'
)
plt.show()
```

```python
# apply function

data_series_2 = np.exp(data_series)

# print

print('series <exponential function>:')
for index in range(len(index_list)):
    print(f'{index_list[index]}{data_series[index]:10.6f}{data_series_2[index]:9.6f}')

# plot

plt.plot(
    data_series, 'o:',
    data_series_2, 'o:'
)
plt.show()
```

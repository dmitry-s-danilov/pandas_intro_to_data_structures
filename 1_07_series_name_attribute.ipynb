# Name attribute

1. **Series** can also have a **name** attribute.
2. The **Series.name** will be assigned automatically in many cases, 
   in particular when taking 1D slices of **DataFrame**.
3. You can rename a Series with the **Series.rename** method.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create series from random ndarray,
# when an index is passed.

i = list('abcdefghijk')
a = np.random.randn(len(i))
s_0 = pd.Series(a, index=i, name='normal 0')

print(s_0)

plt.axhline(y=0, linewidth=1, color='k')
plt.plot(s_0, 'o:')
plt.xlabel(xlabel='label')
plt.ylabel(ylabel='value')
plt.title(s_0.name)
plt.show()
```

```python
# Modify series,
# changing the name

print(s_0, end='\n\n')

# change the name
s_10 = s_0.rename('normal 1')
# remain the name
s_10 = s_10[::2]
# forget the name
s_11 = s_0 - s_10

print(s_10, s_11,
      sep='\n', end='\n\n')

# change the name
s_20 = s_0.rename('normal 2')
# remain the name
s_20 = s_20[1::2]
# forget the name
s_21 = s_0 - s_20

print(s_20, s_21,
      sep='\n', end='\n\n')

plt.axhline(y=0, linewidth=1, color='k')
plt.plot(s_0, '--', label=s_0.name)
plt.plot(s_10, 'o:', label=s_10.name)
plt.plot(s_20, 'o:', label=s_20.name)
plt.xlabel('label')
plt.ylabel('value')
plt.legend()
plt.show()
```

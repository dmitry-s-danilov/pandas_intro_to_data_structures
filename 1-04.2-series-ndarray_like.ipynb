**Series** acts very similarly to a **ndarray**,
and operations such as slicing will also slice the index.


```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create series, when the data is random ndarray,
# and the index is passed.

i = list('abcdefghijk')
a = np.random.randn(len(i))
s = pd.Series(a, i)

plt.plot(s, marker='o', linestyle='dotted')
plt.axhline(y=0, linewidth=1, color='black')
plt.xlabel('label')
plt.ylabel('value')
plt.title('standard normal')
plt.show()
```

```python
# Extract from series
# by random index and random label.

# set random but nonequivalent index and label
while True:
    k = random.choices(population=range(len(s)), k=2)
    if k[0] == k[1]:
        continue
    break
k_1 = k[0]
i_1 = s.index[k[1]]

plt.plot(s,
         marker='o', linestyle='dotted',
         label='random')
# indexing by index
plt.plot(k_1, s[k_1],
         marker='o', linewidth=0,
         label='({},{:>7.3f})'.format(k_1, s[k_1]))
# indexing by label
plt.plot(i_1, s[i_1],
         marker='o', linewidth=0,
         label='({},{:>7.3f})'.format(i_1, s[i_1]))
plt.axhline(y=0,
            linewidth=1, color='black')
plt.xlabel('label')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()  
```

```python
# Extract from series
# by random index list and random label list.

# set random index list
k_1 = random.choices(population=range(len(s)),
                     k=random.randint(a=2, b=len(s)))
# set random label list
i_1 = random.choices(population=s.index,
                     k=random.randint(a=2, b=len(s)))

plt.plot(s,
         marker='o', linestyle='dotted',
         label='random')
# indexing by index list
plt.plot(s[k_1],
         linestyle='dashed',
         label=str(k_1))
# indexing by label list
plt.plot(s[i_1],
         linestyle='dashed',
         label=str(i_1))
plt.axhline(y=0,
            linewidth=0, color='black')
plt.xlabel('label')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

```python
# Slice series by random indexes.

# set a couple of random but growing indexes
while True:
    k_1 = random.choices(population=range(len(s)), k=2)
    if k_1[0] == k_1[1]:
        continue
    elif k_1[1] < k_1[0]:
        k_1 = k_1[::-1]
    break

# slice series by a couple of indexes
s_1 = s[k_1[0]:k_1[1]]

plt.plot(s,
         marker='o', linestyle='dotted',
         label='random')
plt.plot(s_1,
         marker='o', linestyle='solid',
         label=str(k_1))
plt.axhline(y=0,
            linewidth=0, color='black')
plt.xlabel('label')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

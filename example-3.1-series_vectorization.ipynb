```python
# file: example-3.1-series_vectorization.ipynb
# created on 2019-12-11

# pandas Series vectorization examples
    
# when working with Series in pandas,
# looping through value-by-value is usually not necessary,
# as well as when working with raw NumPy arrays
    
# series can also be passed into most NumPy methods
# expecting an ndarray
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# init

# basic way to create a Series is
# to call the pandas.Series() function

array_size = 10 # set ndarray length
data_series_1 = pd.Series(       # create Series with the pandas.Series() function
    np.random.randn(array_size), # providing data by ndarray
                                 # of samples from the "standard normal" distribution and
    name='normal 1'              # setting name attribute
)
data_series_2 = pd.Series(
    np.random.randn(array_size),
    name='normal 2'
)

# print

print(f'i  {data_series_1.name}  {data_series_2.name}')
for index in range(array_size):
    print(
        index,
        f'{data_series_1[index]:10.6f}',
        f'{data_series_2[index]:10.6f}',
        sep=''
    )

# plot

plt.axhline(linewidth=1, color='k')

plt.plot(data_series_1, 'o:', label=data_series_1.name)
plt.plot(data_series_2, 'o:', label=data_series_2.name)

plt.legend()
plt.show()
```

```python
# modify: apply add operator

data_series_3 = data_series_1 + data_series_2
data_series_3 = data_series_3.rename('normal +')

# print

print('i{:>10}{:>10}{:>10}'.format(
        data_series_1.name,
        data_series_2.name,
        data_series_3.name
    )
)
for index in range(array_size):
    print(
        index,
        f'{data_series_1[index]:10.6f}',
        f'{data_series_2[index]:10.6f}',
        f'{data_series_3[index]:10.6f}',
        sep=''
    )

# plot

plt.axhline(linewidth=1, color='k')

plt.plot(data_series_1, '.:', label=data_series_1.name)
plt.plot(data_series_2, '.:', label=data_series_2.name)
plt.plot(data_series_3, 'o-', label=data_series_3.name)

plt.legend()
plt.show()
```

```python
# modify: apply multiply operator

data_series_3 = data_series_1 * data_series_2
data_series_3 = data_series_3.rename('normal *')

# print

print('i{:>10}{:>10}{:>10}'.format(
        data_series_1.name,
        data_series_2.name,
        data_series_3.name
    )
)
for index in range(array_size):
    print(
        index,
        f'{data_series_1[index]:10.6f}',
        f'{data_series_2[index]:10.6f}',
        f'{data_series_3[index]:10.6f}',
        sep=''
    )

# plot

plt.axhline(linewidth=1, color='k')
plt.plot(data_series_1, '.:', label=data_series_1.name)
plt.plot(data_series_2, '.:', label=data_series_2.name)
plt.plot(data_series_3, 'o-', label=data_series_3.name)
plt.legend()
plt.show()
```

```python
# modify: apply NumPy function

data_series_1_e = np.exp(data_series_1)
data_series_1_e = data_series_1_e.rename('exp 1')

# print

print('i{:>10}{:>10}'.format(
        data_series_1.name,
        data_series_1_e.name
    )
)
for index in range(array_size):
    print(
        index,
        f'{data_series_1[index]:10.6f}',
        f'{data_series_1_e[index]:10.6f}',
        sep=''
    )

# plot

plt.axhline(linewidth=1, color='k')
plt.plot(data_series_1, '.:', label=data_series_1.name)
plt.plot(data_series_1_e, 'o-', label=data_series_1_e.name)

plt.legend()
plt.show()
```

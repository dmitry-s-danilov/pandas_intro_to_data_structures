## from dict of ndarrays

1. The **ndarray**s must all be the same length.
2. If an **index** is passed, it must clearly also be the same length as the arrays.
3. If no **index** is passed, the result will be **range(n)**, where **n** is the array length.

```python
import numpy as np
import pandas as pd
```

```python
# Creates a dict of random integers ndarrays,
# whose lengths are equil to the index list length.
i = list('abcdefghij')
a_min, a_max, a_size = 0, 9, len(i)
d = {'one': np.random.randint(low=a_min, high=a_max, size=a_size),
     'two': np.random.randint(low=a_min, high=a_max, size=a_size),
     'three': np.random.randint(low=a_min, high=a_max, size=a_size),
     'four': np.random.randint(low=a_min, high=a_max, size=a_size)}

# Creates a DataFrame from a dict of lists,
# using pandas.DataFrame, when no index is passed.
df = pd.DataFrame(d)

print('dataframe: ', df,
      'index: ', df.index,
      sep='\n', end='\n\n')


# Creates a DataFrame from a dict of lists,
# using pandas.DataFrame, when index is passed.
df = pd.DataFrame(d, index=i)

print('dataframe: ', df,
      'index: ', df.index,
      sep='\n')
```

```python
# Sets a list of columns lists.
c = [['one', 'two', 'three', 'four'],
     ['four', 'three', 'two', 'one'],
     ['one', 'two', 'three'],
     ['two', 'three', 'four'],
     ['one', 'two', 'three', 'four', 'five']]

# Creates a list of DataFrames, using pandas.DataFrame,
# when data is a dict of ndarrays, index and columns are passed.
df = [pd.DataFrame(d, index=i, columns=_)
      for _ in c]

for _ in df:
    print('dataframe:', _,
          'columns:', _.columns,
          sep='\n', end='\n\n')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# init
    # basic way to create a Series is
    # to call the pandas.Series(data, index) function

label_list = list('abcdefgh') # set index label list

array_size = len(label_list)             # set ndarray length
                                         # equal to label list length
data_array = np.random.randn(array_size) # create specified length data ndarray
                                         # of samples from the "standard normal" distribution

data_series = pd.Series( # create Series with the pandas.Series() function
    data_array,          # providing data by ndarray,
    label_list,          # passing index by list and
    name='normal 0'      # setting name attribute
)


# print

print(
    f'{type(data_array)}:', data_array, '',
    f'{type(data_series)}:', data_series,
    sep='\n'
)

# plot

plt.plot(data_series, 'o:') 
plt.show()
```

```python
# modify
    # the pandas.Series.name attribute
    # will be assigned automatically in many cases,
    # in particular when taking slices
    
    # you can rename a Series 
    # with the pandas.Series.rename() method

data_series_10 = data_series.rename('normal 1') # change
data_series_10 = data_series_10[::2]            # remain
data_series_11 = data_series - data_series_10   # clear

data_series_20 = data_series.rename('normal 2') # change
data_series_20 = data_series_20[1::2]           # remain
data_series_21 = data_series - data_series_20   # clear

# print

print(
    '0:', data_series, '',
    '1.0:', data_series_10, '',
    '1.1:', data_series_11, '',
    '2.0:', data_series_20, '',
    '2.1:', data_series_21, 
    sep='\n'
)

# plot

plt.plot(data_series, '--', label=data_series.name)
plt.plot(data_series_10, 'o:', label=data_series_10.name)
plt.plot(data_series_20, 'o:', label=data_series_20.name)
plt.legend()
plt.show()
```

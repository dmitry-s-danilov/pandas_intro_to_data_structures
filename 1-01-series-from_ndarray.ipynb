1. The basic method to create a **Series** is
   to call **pandas.Series**(*data*, *index*),
   where *data* can be an **ndarray**,
   and the passed *index* is a list of axis labels.
2. If *data* is an ndarray, and no *index* is passed,
   one will be created having values 0, ..., len(data) - 1.
3. If *data* is an ndarray, *index* must be the same length as *data*.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create series, when the data is an ndarray,
# and no index is passed.

a = np.random.rand(10 + 1)
s = pd.Series(data=a)

print(f'data {type(a)}:', a, '',
      f'series {type(s)}:', s,
      sep='\n')

plt.plot(s, marker='o', linestyle='dotted')
plt.xticks(s.index)
plt.xlabel('index')
plt.ylabel('value')
plt.title('random')
plt.show()
```

```python
# Create series, when the data is an ndarray, 
# and the same length index is passed.

i = list('abcdefghijk')
a = np.random.rand(len(i))
s = pd.Series(a, i)

print(f'data {type(a)}:', a, '',
      f'index {type(i)}:', i, '',
      f'Series {type(s)}:', s,
      sep='\n')

plt.plot(s, marker='o', linestyle='dotted')
plt.xlabel('label')
plt.ylabel('value')
plt.title('random')
plt.show()
```

```python
# Create series, when the data is an ndarray, 
# and not the same length index is passed.

i = list('abc')
a = np.random.rand(len(i) + 1)
s = pd.Series(a, i)
```

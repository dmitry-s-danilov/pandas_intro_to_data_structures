```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create series

index_list = list("abcdefgh") # set index list

array_length = len(index_list)             # set array length and
data_array = np.random.randn(array_length) # initialize data ndarray

data_series = pd.Series(data_array, index_list) # create data series
                                                # from data ndarray providing index list

# print

print(
    f'index {type(index_list)}:',
    f'{index_list}',
    '',    
    f'data {type(data_array)}:',
    f'{data_array}',
    '',    
    f'series {type(data_series)}:',
    f'{data_series}',
    sep='\n'
)

# plot

plt.plot(data_series, 'o:')
plt.show()
```

```python
# extract by single index

index_1 = random.choice(range(len(index_list))) # set random index
                                                # within the limits of initial index list 
data_series_1 = data_series[index_1]            # extract single value
                                                # from initial series by single index

# print

print('series <indexing>:')
for index, value in enumerate(data_series):
    print(f'{index_list[index]}{value:10.6f}', end='')
    if index == index_1:
        print(' *')
    else:
        print()

# plot

plt.plot(
    data_series, '.:',
    index_1, data_series_1, 'o'
)
plt.show()
```

```python
# extract by index list

# set index list length
# exceeding by unit initial index list length
# to ensure duplicate indexes appearance
index_list_2_length = len(index_list) + 1

# generate random index list
# with duplicate indexes
index_list_2 = random.choices( 
        population=range(len(index_list)),
        k=index_list_2_length
)

# extract series
# from initial series by index list
data_series_2 = data_series[index_list_2]

# print

print(
    "series <indexing>:",
    data_series_2,
    sep='\n'
)

# plot

plt.plot(
    data_series, '.:',
    data_series_2, 'o-'
)
plt.show()
```

```python
# slice

# set random range list
while True:
    # generate random range list
    # within the limits of initial index list
    slice_range_3 = random.choices( 
        population=range(len(index_list)),
        k=2
    )

    # append step value to range list
    if slice_range_3[0] == slice_range_3[1]:
        continue
    elif slice_range_3[0] < slice_range_3[1]:
        slice_range_3_step = 1
    else:
        slice_range_3_step = -1
    slice_range_3.append(slice_range_3_step)
    break

# extract series
# from initial series by slice
# using range list values
data_series_3 = data_series[
    slice_range_3[0]:
    slice_range_3[1]:
    slice_range_3[2]    
]

# print

print(
    'series <slicing>:',
    data_series_3,
    sep='\n'
)

# plot

plt.plot(
    data_series, '.:',
    data_series_3, 'o-'
)
plt.show()
```

# Series from ndarray

- The basic method to create a `Series` is to call: 

  ```python
  pandas.Series(data, index)
  ```

  where `data` can be an `ndarray`,
  and the passed `index` is a `list` of axis labels.

- If `data` is an ndarray,
  `index` must be the same length as `data`.

- If `data` is an ndarray, and no `index` is passed,
  one will be created having values `[0, ..., len(data) - 1]`.

- Pandas supports non-unique index values.
  If an operation that does not support duplicate index values is attempted,
  an exception will be raised at that time.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Creates series 
# when the data is an array and
# no index is passed.

a = np.random.rand(10 + 1)
s = pd.Series(data=a)

print(f'data {type(a)}:', a, '',
      f'series {type(s)}:', s,
      sep='\n')

# plt.plot(s, marker='o', linestyle='dotted')
plt.plot(s, 'o:')
plt.xticks(s.index)
plt.xlabel('index')
plt.ylabel('data')
plt.title('series')
plt.show()
````

````python
# Creates series when
# the data is an array and
# the same length index is passed.

a = np.random.rand(10)
i = [chr(ord('a') + _) for _ in range(a.size)]
s = pd.Series(a, i)

print(f'data {type(a)}:', a, '',
      f'index {type(i)}:', i, '',
      f'Series {type(s)}:', s,
      sep='\n')

plt.plot(s, 'o:')
plt.xticks(s.index)
plt.xlabel('index')
plt.ylabel('data')
plt.title('series')
plt.show()
````

````python
# Creates series when
# the data is an array and
# the same length non-unique index is passed.

a = np.random.rand(10)
i = np.random.choice(
    [chr(ord('a') + _) for _ in range(a.size - 1)],
    size=a.size)
s = pd.Series(a, i)

print(f'data {type(a)}:', a, '',
      f'index {type(i)}:', i, '',
      f'Series {type(s)}:', s,
      sep='\n')

plt.plot(s, 'o:')
plt.xticks(s.index)
plt.xlabel('index')
plt.ylabel('data')
plt.title('series')
plt.show()
````

````python
# Create series when
# the data is an ndarray and
# not the same length index is passed.

a = np.random.rand(10)
i = range(a.size - 1)
s = pd.Series(a, i)
````

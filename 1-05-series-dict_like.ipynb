1. A Series is like a fixed-size dict
   in that you can get and set values by index label.
2. If a label is not contained, an exception is raised.
3. Using the **Series.get** method,
   a missing label will return None or specified default.


```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create series from random ndarray,
# when an index is passed.

i = list('abcdefghijk')
a = np.random.rand(len(i))
s = pd.Series(a, index=i, name='random')

print(s)

plt.plot(s, 'o:')
plt.ylim(ymin=0)
plt.xlabel(xlabel='label')
plt.ylabel(ylabel='value')
plt.title(s.name)
plt.show()
```

```python
# Get a value from the series
# by rendom index label.

k = random.choice(s.index)

print('s:', s, '',
      f'{k} in s: {k in s}',
      f's[{k}]: {s[k]:.6f}',
      f's.get({k}): {s.get(k):.6f}',
      sep='\n')

plt.plot(s, '.:', label=s.name)
plt.plot(k, s[k], 'o', label=f'({k}, {s[k]:.6f})')

plt.ylim(ymin=0)
plt.xlabel(xlabel='label')
plt.ylabel(ylabel='value')
plt.legend()
plt.show()
```

```python
# Try to get a value from the series
# by rendom index label
# that is not contained.

k = 'x'

print('s:', s, '',
      f'{k} in s: {k in s}',
      # an exception is raised
      # f's[{k}]: {s[k]:.6f}',
      f's.get({k}): {s.get(k)}',
      sep='\n')
```
